```python
import requests
import bs4
```

```python
baseUrl="https://gogoanime2.org/"
res = requests.get(baseUrl)
```

```python
soup=bs4.BeautifulSoup(res.text,'lxml')
```

```python
pageTitle=soup.select('title')[0].getText()
```

```python
pageTitle
```

```python
#select all img tags inside items class, go to web dev course for this
listItems=soup.select('.items img')
```

```python
listItems
```

```python
for imgTags in listItems:
    print(imgTags['alt'])
    print(baseUrl+imgTags['src'])
```

```python
#pagination
nextPages=soup.select('.pagination-list li a')
```

```python
nextPages
```

```python
for nextPagesTags in nextPages:
    print(baseUrl+nextPagesTags['href'])
```

```python
#writing a loop to fetch all anime Names using pagination
import requests
import bs4
count=0
baseUrl="https://gogoanime2.org/"
nextPage='/ajax/page-recent-release.html?page=4'
#max can be 125
while(count<=10):
    try:
        res = requests.get(baseUrl+nextPage)
        count+=1
        soup=bs4.BeautifulSoup(res.text,'lxml')
        listItems=soup.select('.items img')
        for imgTags in listItems:
            print(imgTags['alt'])
            print(baseUrl+imgTags['src'])
        print('\nNEXT-PAGE\n')
        nextPage=soup.select('.pagination-list li a')[4]['href']
    except:
        break
    
    
```
